```python
# Retrieve information from Twitter and store in pickle file
import tweepy
from twitter_keys import consumer_key, consumer_secret, access_token, access_secret
import pickle

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)

tarak9999_tweets = api.user_timeline('tarak9999')

with open('tarak9999_tweets.pkl', 'wb') as f:
    pickle.dump(tarak9999_tweets, f)
    
RanbirKUniverse_tweets = api.user_timeline('RanbirKUniverse')

with open('RanbirKUniverse_tweets.pkl', 'wb') as f:
    pickle.dump(RanbirKUniverse_tweets, f)
    
    
imVkohli_tweets = api.user_timeline('imVkohli')

with open('imVkohli9_tweets.pkl', 'wb') as f:
    pickle.dump(imVkohli_tweets, f)
```

```python
# Load cached data from pickle file
import pickle

with open('tarak9999_tweets.pkl', 'rb') as f:
    tarak9999_tweets = pickle.load(f)

print('tarak tweets',tarak9999_tweets)

with open('RanbirKUniverse_tweets.pkl', 'rb') as f:
    RanbirKUniverse_tweets = pickle.load(f)

print('ranbir tweets',RanbirKUniverse_tweets)

with open('imVkohli_tweets.pkl', 'rb') as f:
    imVkohli_tweets = pickle.load(f)

print('kohli tweets ',imVkohli_tweets)
```

# Description of analysis steps


1. In the below cells analysis of the three twitter accounts have been done using lexical diversity and sentiment analysis.
1. In the analysis using lexical diversity, I have
    - retrived the first tweet of all the three people
    - obtained the tokens in the firts tweet
    - calculated the length of the first tweet
    - printed the tokens in lower case
    - printed the unique tokens in the first tweet
    - finally calculated and printed the lexical diversity of the first tweet using the formula, (length of the unique tokens in lower) / (length of the tokens in lower case)
1. In the sentiment analysis, I found how much positive or negative or neutral the different tweets of each person are, and printed the results for all the three persons. I then plotted the results on a graph.

```python
# Lexical Diversity Analysis 

# First tweet of twitter account 1
tweet = tarak9999_tweets[0]
print('First tweet of tarak ',tweet.text)

# Tokens in first tweet of twitter account 1
tokens = tweet.text.split()
print(tokens)

# Length of twitter account 1
print(len(tokens))

# Tokens in lower case, unique tokens and lexical diversity of twitter account 1
lower_tokens = [s.lower() for s in tokens]
print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity OF twitter account 1:', len(unique_lower)/len(lower_tokens))

# First tweet of twitter account 2
tweet1 = RanbirKUniverse_tweets[0]
print('First tweet of ranbir',tweet1.text)

# Tokens in first tweet of twitter account 2
tokens = tweet1.text.split()
print(tokens)

# Length of twitter account 2
print(len(tokens))

# Tokens in lower case, unique tokens and lexical diversity of twitter account 2
lower_tokens = [s.lower() for s in tokens]
print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity OF twitter account 2:', len(unique_lower)/len(lower_tokens))

# First tweet of twitter account 3
tweet2 = imVkohli_tweets[0]
print('First tweet of kohli ',tweet2.text)

# Tokens in first tweet of twitter account 3
tokens = tweet2.text.split()
print(tokens)

# Length of twitter account 3
print(len(tokens))

# Tokens in lower case, unique tokens and lexical diversity of twitter account 3aQ
lower_tokens = [s.lower() for s in tokens]
print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity OF twitter account 3:', len(unique_lower)/len(lower_tokens))

      
```

```python
Sentiment Analysis

import pickle
from nltk_helpers import get_sentiments

# Sentiment Analysis for the first twitter account - tarak9999

print('Tarak\'s tweets analysis:')
for tweet in tarak9999_tweets:
    print(get_sentiments(tweet.text))

# Sentiment Analysis for the first twitter account - Ranbir Kapoor

print('Ranbir\'s tweets analysis:')
for tweet in RanbirKUniverse_tweets:
    print(get_sentiments(tweet.text))
    
# Sentiment Analysis for the first twitter account - imVkohli

print('Virat Kohli\'s tweets analysis:')
for tweet in imVkohli_tweets:
    print(get_sentiments(tweet.text))
```

```python
# Plotting the reaults of twitter account 1
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# build our figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tarak9999_tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('arbys.pkl', 'rb') as f:
    arbys_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

arbyX, arbyY, arbyZ = split_sentiments([get_sentiments(tweet.text) for tweet in arbys_tweets])
ax.scatter(arbyX, arbyY, arbyZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')

plt.show()
```
